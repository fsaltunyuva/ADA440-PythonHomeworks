```python
def M(n:int) -> int:
    # Your code here!
    unique_values = []
    for i in range(1,n+1):
      for j in range(1,n+1):
        multiplication = i*j
        if multiplication not in unique_values:
            unique_values.append(i*j)
    return len(unique_values)
```

# Primitive Gauss Circle Problem:
# Given $r > 0$, you will write a function that gives the number of all $(x,y) \in \mathbb{Z}^2$ and $x^2 + y^2 \le r^2$ such that $x,y$ have no common divisors  (Use math.gcd function).

```python
def primitive_circle(r:int)->int:
    ## Your code here
    x = -r
    y = -r
    result = 0

    for i in range(x, r+1):
      for j in range(y, r+1):
        if math.gcd(i,j) == 1 and i**2 + j**2 <= r**2:
          result += 1

    return result
```

```python
### Below you will implement a function that returns lower cased token frequencies in a dictionary,
### feel free to code as you like,
def token_freq(text:str)->dict:
    ## your code here!!!
    non_uniques = text.lower().split()
    result = {}

    for i in non_uniques:
      if i in result:
        result[i] += 1
      else:
        result[i] = 1

    return result
```
